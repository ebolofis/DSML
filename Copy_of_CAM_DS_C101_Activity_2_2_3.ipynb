```python
# Import libraries
import pandas  as pd #Data manipulation
import numpy as np #Data manipulation
import matplotlib.pyplot as plt #Visualisation
import seaborn as sns #Visualisation
from scipy import stats # place the import at the top of the file.


data = pd.read_csv("https://raw.githubusercontent.com/fourthrevlxd/cam_dsb/main/insurance_updated.csv")
# Correlation and p-values
def calculate_correlation_and_pvalue(data, col1, col2):
    corr, p_value = stats.pearsonr(data[col1], data[col2])
    print(f"Correlation between {col1} and {col2}: {corr}, p-value: {p_value}")
# Descriptive Statistics
print(data.describe())

# Check for missing values
print(data.isnull().sum())

# Histograms for numerical variables
numerical_cols = ['age', 'bmi', 'number_of_children', 'charges']
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.histplot(data[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()

# Scatter plots to visualize relationships
sns.pairplot(data)
plt.show()

# Box plots to check for outliers
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=data[col])
    plt.title(f'Box plot of {col}')
    plt.show()

    # Correlation Matrix (Pearson)
correlation_matrix = data[numerical_cols].corr(method='pearson')
print("Pearson Correlation Matrix:\n", correlation_matrix)

# Heatmap for visualization
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Pearson Correlation Heatmap')
plt.show()

calculate_correlation_and_pvalue(data, 'age', 'charges')
calculate_correlation_and_pvalue(data, 'bmi', 'charges')
calculate_correlation_and_pvalue(data, 'number_of_children', 'charges')
```
